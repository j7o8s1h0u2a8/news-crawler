```python
#####
##### 爬內容的方法
def crawlPage(url):
    
    res = requests.get(url)
    
    soup = BeautifulSoup(res.text,'lxml')

    ## 去除非文字部分
        #player-and-links
        #infgraphicsAttach
    for push in soup.select("span"):
        push.extract()
    
    for push in soup.select("div.wsw > div"):
        push.extract()
      
    contents = soup.select('div.wsw > p')

    articles_dict = {}

    for page in range(len(contents)+len(str(contents).split('<br/>\n<br/>'))):
    
        articles_dict[page+1] = contents[page].text
    
    return articles_dict
```

```python
crawlPage('http://www.voanews.com/a/somalia-corruption/3792202.html')
```
